```python
import json

import numpy as np

import visualization as vis

help(vis.show)
```

```python
# coco prediction loading
D = json.load(open("pred/coco_box.json","r"))

for i in range(3):
    # prepare data
    print(D[f"pic{i}.jpg"])
    pd = np.array(D[f"pic{i}.jpg"]["preds"])
    pd_cids, pd_cfs = pd.argmax(axis=1), pd.max(axis=1)
    
    # visualization
    vis.show(["dog","cat"], f"data/coco/pic{i}.jpg", "data/coco/coco.json", \
        "coco", D[f"pic{i}.jpg"]["boxes"], pd_cids, pd_cfs)
```
