```python
import matplotlib.pyplot as plt
import re, json

txt = open("DETReg_top30_in/log.txt","r").read()
txtL = re.findall("\{(.*)\}", txt) # epoch split
train_loss_L = [ json.loads("{"+txtL[i]+"}")["train_loss"] for i in range(len(txtL)) ]
plt.title("Pretask on recycling data", fontsize=16)
plt.xlabel("epochs", fontsize=16)
plt.ylabel("loss", fontsize=16)
plt.plot(train_loss_L[10:])
plt.grid('on')
plt.show()
```

```python
txt = open("DETReg_fine_tune_full_coco//log.txt","r").read()
txtL = re.findall("\{(.*)\}", txt) # epoch split
train_loss_L = [ json.loads("{"+txtL[i]+"}")["train_loss"] for i in range(len(txtL)) ]
valid_loss_L = [ json.loads("{"+txtL[i]+"}")["test_loss"] for i in range(len(txtL)) ]
plt.title("Finetune on recycling data", fontsize=16)
plt.xlabel("epochs", fontsize=16)
plt.ylabel("loss", fontsize=16)
plt.plot(train_loss_L)
plt.plot(valid_loss_L)
plt.grid('on')
plt.show()
```
