### 1. show_coco

```python
help(show_coco)
```

```python
for i in range(3):
    show_coco(
        img_name = f"pic{i}.jpg",
        img_folder = "example/detection/data/coco",
        ant_path = "example/detection/data/coco/coco.json",
        use_cache = False
    )
```

### 2. show_voc

```python
help(show_voc)
```

```python
for i in range(3):
    show_voc(
        img_name = f"pic{i}.jpg",
        img_path_list = sorted(glob.glob("example/detection/data/voc/*.jpg")),
        ant_path_list = sorted(glob.glob("example/detection/data/voc/*.xml")),
        class_list = ["dog", "cat"],
        use_cache = False
    )
```

### 3. show_yolo

```python
help(show_yolo)
```

```python
for i in range(3):
    show_yolo(
        img_name = f"pic{i}.jpg",
        img_path_list = sorted(glob.glob("example/detection/data/yolo/*.jpg")),
        ant_path_list = sorted(path for path in glob.glob("example/detection/data/yolo/*.txt") if "classes.txt" not in path),
        class_list = ["dog", "cat"],
        use_cache = False
    )
```
